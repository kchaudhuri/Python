```python
def isprime(n):
    if n == 1:
        return False
    for x in range(2,n):
        if n % x == 0:
            return False
        else:
            return True
```

```python
def primes(n=1):
    while(True):
        if isprime(n):
            yield n
        n += 1
```

```python
for n in primes():
    if n>100:
        break
    print(n)
```

```python
class sclass():
    def __init__(self,a,b):
        self.a = a
        self.b = b
        
    def sfunc(self):
        print(self.a+self.b)
        
sobj = sclass(0,1)
sobj.sfunc()
```

```python
class Fibonacci():
    def __init__(self,a,b):
        self.a = a
        self.b = b
        
    def series(self):
        while(True):
            yield(self.b)
            self.a, self.b = self.b, self.a+self.b
            
f = Fibonacci(0,1)
for r in f.series():
    if r > 100:
        break
    print(r)
```

```python
def gen():
    a = 0
    while a < 100:
        yield(a)
        a = a + 1
        yield(a)
```

```python
for i in gen():
    print(i)
```
